```python
import numpy as np
import pandas as pd
import glob
```

```python
path = '../data/'
documents = ['photos', 'keywords', 'collections', 'conversions', 'colors']
datasets = {}

for doc in documents:
  files = glob.glob(path + doc + ".tsv*")

  subsets = []
  for filename in files:
    df = pd.read_csv(filename, sep='\t', header=0)
    subsets.append(df)

  datasets[doc] = pd.concat(subsets, axis=0, ignore_index=True)
```

```python
for req_key in datasets.keys():
    print(f"Dataset: {req_key}, Shape: {datasets[req_key].shape}")
    print(f"Columns: {datasets[req_key].columns.tolist()}")
    display(datasets[req_key].head())
    print("\n")
```

```python
# save as parquet in data folder
for req_key in datasets.keys():
    datasets[req_key].to_parquet(f"../data/metadata/{req_key}.parquet", index=False)
```
