```python
from huggingface_hub import HfApi
import os
```

```python
hf_token = os.environ.get("MOBILECLIP_APP_TOKEN")
api = HfApi(token=hf_token)
repo_id = "devanshu125/unsplash-lite-small"
```

```python
info = api.dataset_info(repo_id)
print(info)
print(f"Number of files: {len(info.siblings)}")
print(f"Last modified: {info.last_modified}")
```

```python
# count number of files in the images folder
files = [s for s in info.siblings if s.rfilename.startswith("data/images/")]
print(f"Number of image files: {len(files)}")
print(f"First 10 image files:")
files[:10]
```
